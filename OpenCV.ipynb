First code of opencv in colab to show an image. Also using matlplotlib.pyplot to show the image.

```python
import cv2 as cv
print(cv.__version__)
import sys
import matplotlib.pyplot as plt
```

```python
img = cv.imread(cv.samples.findFile("/content/drive/MyDrive/cat.jpg"))
```

```python
if img is None:
  sys.exit("image not found")
```

```python
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.show()
```

Now lets learn to crop the image.

```python
upper = 1000  # value set to upper for the upper pixel
lower = 3500   # value set to lower for the lower pixel
img_crop_height = img[upper:lower,:]
plt.imshow(cv.cvtColor(img_crop_height,cv.COLOR_RGB2BGR))
plt.title("Height Croped Image")
plt.show()
```

Now after cropping the height we will crop the length.

```python
left = 500     #left pixel is defined
right = 2600   #right pixel is defined
img_crop_weidth = img_crop_height[:,left:right]
plt.imshow(cv.cvtColor(img_crop_weidth,cv.COLOR_RGB2BGR))
plt.title("Weidth Cropped Image")
plt.show()
```
