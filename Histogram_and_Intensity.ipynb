we will learn to change the brightness and contrast i.e intensity of an image. 

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread("/content/drive/MyDrive/cat.jpg")
plt.imshow(cv.cvtColor(img,cv.COLOR_BGR2RGB))
plt.title("Original Pic")
plt.show()
cat = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(cat,cmap="gray")
plt.title("Gray Pic")
plt.show()
```

```python
hist1 = cv.calcHist([img],[0],None,[256],[0,256])
hist2 = cv.calcHist([img[:,:1000]],[0],None,[256],[0,256])
plt.plot(hist1)
plt.xlim([0,256])
plt.show()
plt.plot(hist2)
plt.show()
```

histogram of RGB.


```python
color = {"blue","green","red"}
for index,col in enumerate(color):
  hist = cv.calcHist([img],[index],None,[256],[0,256])
  plt.plot(hist,color = col,label = col)
  # plt.color(col)
  # plt.lable(color)
  plt.xlim([0,256])
plt.show()
```

Image negative we have the formula [(new intensity) = 255- (old intensity)]

This will make the brighter to darker and vice-versa. Middle intensiy will remain same.

```python
new_cat = -1*cat +255
plt.imshow(cat, cmap="gray")
plt.title("original gray cat") 
plt.show()
plt.imshow(new_cat, cmap= "gray")
plt.title("nagative image")
plt.show()
```
